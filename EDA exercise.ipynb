Características técnicas y de identificación de diferentes modelos de automóviles:   

- **mpg**: Millas por galón (Miles Per Gallon), una medida de la eficiencia de combustible del automóvil, que indica cuántas millas puede recorrer el automóvil por cada galón de combustible utilizado.

- **cylinders**: Número de cilindros en el motor del automóvil. Es una variable discreta que indica la cantidad de cilindros presentes en el motor del automóvil, que puede ser 4, 6 u 8 en la mayoría de los casos.

- **displacement**: Desplazamiento del motor, una medida del volumen total de los cilindros del motor. Se mide típicamente en litros o centímetros cúbicos y representa la capacidad del motor para generar potencia.

- **horsepower**: Potencia del motor, que representa la cantidad de trabajo que el motor puede realizar en una unidad de tiempo. Es una medida de la capacidad del motor para realizar trabajo mecánico y suele expresarse en caballos de fuerza (HP).

- **weight**: Peso del automóvil, que indica la masa total del vehículo, incluyendo todos sus componentes y carga. Se mide típicamente en libras o kilogramos.

- **acceleration**: Aceleración del automóvil, que representa la capacidad del vehículo para aumentar su velocidad en un período de tiempo determinado. Se mide típicamente en segundos para acelerar de 0 a 60 millas por hora (mph) o de 0 a 100 kilómetros por hora (km/h).

- **model_year**: Año del modelo del automóvil, que indica el año de fabricación del vehículo.

- **origin**: Origen del automóvil, que indica el país de origen o fabricación del vehículo.

- **name**: Nombre del modelo del automóvil, que proporciona información sobre el modelo específico del vehículo.

```python
plt.figure(figsize=(10,8))
heatmap = sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidths=0.5, vmax=1, center=0, square=True)           
heatmap.set_title("Correlación entre las variables del dataset")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="origin", y="weight")
plt.title("Relación entre el origen de fabricación y el peso de los vehículos")
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.relplot(x="horsepower", y="weight", data=df, kind="scatter")
sns.kdeplot(x="horsepower", y="weight", data=df, color="r", levels=5, linewidths=1.5, alpha=0.5)
plt.title("Relación entre la potencia del motor y su peso")
plt.xlabel("horsepower")
plt.ylabel("weight")
plt.show()
```
